```python
import pandas as pd

# Load the uploaded file (in Colab path)
df = pd.read_csv('semeval_toxic_spans.csv')

# Display column names
print("Columns:", df.columns.tolist())

# Preview the first few rows
df.head()
```

```python
# Keep only necessary columns
df = df[['text_of_post', 'toxic']]

# Rename columns for consistency
df.columns = ['text', 'label']

# Drop rows with missing values
df.dropna(inplace=True)

# Preview
df.head()
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from transformers import DataCollatorWithPadding
from torch.utils.data import DataLoader, Dataset as TorchDataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
from datasets import Dataset
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Convert pandas dataframe to HuggingFace Dataset
dataset = Dataset.from_pandas(df)

# Load tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, padding="max_length", max_length=128)


# Tokenize the dataset
tokenized_dataset = dataset.map(tokenize_function, batched=True)
```

```python
class ToxicCommentsDataset(TorchDataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        return {
            key: torch.tensor(val[idx]) for key, val in self.encodings.items()
        } | {'labels': torch.tensor(self.labels[idx])}

    def __len__(self):
        return len(self.labels)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
class ToxicDataset(TorchDataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {
            key: torch.tensor(val[idx])
            for key, val in self.encodings.items()
        }
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

# Extract input and labels
# ✅ Updated to match your column names
labels = dataset["label"]
input_ids = tokenized_dataset["input_ids"]
attention_mask = tokenized_dataset["attention_mask"]

# Prepare folds
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

fold_results = []
conf_matrices = []

for fold, (train_idx, val_idx) in enumerate(skf.split(input_ids, labels)):
    print(f"\n🌀 Fold {fold+1}/5 - Progress: {((fold+1)/5)*100:.2f}%")

    # Split inputs
    train_encodings = {
        "input_ids": [input_ids[i] for i in train_idx],
        "attention_mask": [attention_mask[i] for i in train_idx]
    }
    val_encodings = {
        "input_ids": [input_ids[i] for i in val_idx],
        "attention_mask": [attention_mask[i] for i in val_idx]
    }
    train_labels = [labels[i] for i in train_idx]
    val_labels = [labels[i] for i in val_idx]

    train_dataset = ToxicDataset(train_encodings, train_labels)
    val_dataset = ToxicDataset(val_encodings, val_labels)

    print(f"Train samples: {len(train_dataset)} | Validation samples: {len(val_dataset)}")
```

```python
!pip install transformers datasets --upgrade -q # Upgrade transformers and datasets to latest version

from transformers import BertForSequenceClassification, TrainingArguments, Trainer
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, precision_recall_fscore_support
from datasets import Dataset

# Function to compute metrics
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='binary')
    acc = accuracy_score(labels, preds)
    return {
        'accuracy': acc,
        'precision': precision,
        'recall': recall,
        'f1': f1,
    }

fold_results = []

for fold, (train_idx, val_idx) in enumerate(skf.split(tokenized_dataset, labels)):
    print(f"\n🔄 Fold {fold + 1}/5")

    # Split dataset
    train_dataset = tokenized_dataset.select(train_idx)
    val_dataset = tokenized_dataset.select(val_idx)

    # Define model
    model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

    # Training arguments
    training_args = TrainingArguments(
        output_dir=f"./results_fold{fold}",
        num_train_epochs=3,              # ✅ Set to 3 epochs
        per_device_train_batch_size=16,
        per_device_eval_batch_size=64,
        # Replace evaluation_strategy with the correct argument if necessary:
        # evaluation_strategy="epoch",
        save_strategy="no", # "epoch" to save at the end of each epoch
        logging_dir=f"./logs_fold{fold}",
        learning_rate=2e-5,
        load_best_model_at_end=False, # set to True to automatically load the best model
        seed=42, # random seed for reproducibility
        disable_tqdm=False, # whether to disable the tqdm progress bar
        logging_steps=50, # how often to log training metrics
        report_to="none" #  where to report metrics (e.g., "wandb", "tensorboard")
        # ... other arguments ...
    )

    # Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        compute_metrics=compute_metrics
    )

    # Train and evaluate
    trainer.train()
    eval_results = trainer.evaluate()
    fold_results.append(eval_results)

# Display all fold results
for i, result in enumerate(fold_results):
    print(f"\n📊 Fold {i+1} Metrics:")
    for metric, value in result.items():
        print(f"{metric}: {value:.4f}")
```

```python
# Step 4: Summarize Fold Metrics
import pandas as pd

# Convert list of dicts into a DataFrame
metrics_df = pd.DataFrame(fold_results)

# Show average metrics across all folds
average_metrics = metrics_df.mean()
print("\n🔍 Average Metrics Across 5 Folds:")
for metric, value in average_metrics.items():
    print(f"{metric}: {value:.4f}")
```

```python
# Step 7: Train on full dataset
final_model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

final_trainer = Trainer(
    model=final_model,
    args=training_args,
    train_dataset=tokenized_dataset,
    eval_dataset=tokenized_dataset,
    compute_metrics=compute_metrics,
)

# Train
final_trainer.train()
```

```python
import torch

# Make sure to move the model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
final_model.to(device)

def predict_toxicity(text):
    # Tokenize and move inputs to the model's device
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)

    # Run the model
    outputs = final_model(**inputs)

    # Get softmax probability of toxic class
    probs = torch.nn.functional.softmax(outputs.logits, dim=1)
    toxicity_score = probs[0][1].item()  # index 1 = toxic class
    return toxicity_score
```

```python
def predict_toxicity(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)
    outputs = model(**inputs)
    probs = torch.softmax(outputs.logits, dim=1)
    toxicity_score = probs[0][1].item()

    # Label based on score range
    if toxicity_score <= 0.30:
        level = "✅ Safe"
        print("It is Safe")
    elif toxicity_score <= 0.60:
        level = "⚠️ Less Toxic"
        print("It is Less Toxic")
    else:
        level = "❌ Toxic"
        print("It is Toxic")

    return toxicity_score, level

# Example usage
text = "You are good girl!"
score, level = predict_toxicity(text)
print(f"Toxicity Score: {score:.2f} | Level: {level}")
```
